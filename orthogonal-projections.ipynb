## Exercise 7: Write a Python Script to solve Example 3.11, using numpy and to verify the results

```python
# Initialization of the basis matrix and the given vector 
B=np.array([
    [1,0],
    [1,1],
    [1,2]
])
x=np.array([6,0,0])
# if the rank is not full, then we break the program as otherwise we don't have a basis
if np.linalg.matrix_rank(B) != 2:
     raise Exception("Not Full Rank")
# follow the steps of the book to compute the projection
BTB=B.T @ B
BTx=B.T @ x
# coefficients
lamCoeff=np.linalg.solve(BTB, BTx)
# orthogonal projection
proj = B @ lamCoeff
print(proj)
# matrix

```
