```python
## pth Root of x
def pRoot(x,p):
    return x**(1/p)
    
## lp Norm
def lpNorm(l,p):
    sums=0
    for x in l:
        sums+=abs(x)**p
    return pRoot(sums,p)

## compare max value of l with lp as p grows
def convergenceTest(l, p, tolerance):
    maxVal=max(l)
    while abs(lpNorm(l,p)-maxVal) > tolerance:
        p+=1
        print(f"p={p}, lpNorm={lpNorm(l,p):.6f}")
    return p, lpNorm(l,p)
```
