```python
import tensorflow as tf
import numpy as np
```

```python
def f1(x):
  return tf.sqrt(x**2+tf.exp(x**2))+tf.cos(x**2+tf.exp(x**2))
x=tf.Variable(3.0)
with tf.GradientTape() as tape:
    y = f1(x)
grad = tape.gradient(y, x)

print("Function value:", y.numpy())
print("Gradient:", grad.numpy())
```

```python
def f2(x):
  return 2*x * (1 + tf.exp(x**2)) * ((2*tf.sqrt(x**2 + tf.exp(x**2)))**-1 - tf.sin(x**2 + tf.exp(x**2)))
print(f2(3.0))

```
