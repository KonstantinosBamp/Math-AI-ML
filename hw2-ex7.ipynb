```python
import numpy as np
```

```python
## define the vectors

x1 = np.array([1,1,1])
x2 = np.array([1,2,3])
x3 = np.array([2,-1,1])

y = np.array([1,-2,5])

## create the matrix 

A = np.column_stack((x1, x2, x3))

## use linalg solver 

coefficients = np.linalg.solve(A, y)

l1,l2,l3 = coefficients 

print("l1 =", l1, "l2 =", l2, "l3 =", l3)
```

```python
## check that the solution is correct

y_Check = (l1*x1) + (l2*x2) + (l3*x3)
print(y_Check)
```
